# 문자열에서 자주 사용하는 함수

```python
a = "Life is too short!"
print(f"문자(열) 개수 세기:  {a.count('o')}")
print(f"문자(열) 위치: {a.find('f')}")  # 없으면 '-1' 반환
print(f"문자(열) 위치: {a.index('f')}") # 없으면 오류
b = '*'
print(f"문자열 삽입: {b.join(a)}")        # a에 b를 삽입
print(f"문자열 변환: {a.replace('Life','Your leg')}")  # 문자열 대체
print(f"문자열 나누기: {a.split()}")      # 공백으로 분리해서 리스트형으로 반환
print(f"문자열 나누기: {a.split('is')}")
```

### 파이썬 연산자

```python
# 산술 연산자
print(33 // 5) # 몫
print(33 % 5) # 나머지
```

```python
# 대입 연산자
a = a + 1 # a += 1
meal = meal + (meal * tip)
meal += (meal * tip)
```

```python
# 관계 연산자
a = 100
print ( a > 200)
print ( a == 100)
print ( a != 100)
```

```python
# 논리 연산자
a = 100
print (a > 100 and a == 100)
print (a > 100 or a == 100)
print ( not(a==100) )
```

```python
# 멤버쉽 연산자
fruits = ['사과','포도','오렌지','딸기']
print('딸기' in fruits)
print('키위' not in fruits)

if '딸기' in fruits:
  print('딸기는 fruits 안에 있습니다')
else:
  print('딸기는 fruits 안에 없습니다')
```

```python
for fruit in fruits:
  print(f'내가 좋아하는 과일 "{fruit}"')
```

```python
coin = int(input('동전으로 교환할 금액을 입력하세요: '))
coins = coin // 500
coins1 = coin % 500
print("-" * 30)
print(f"500원짜리 ==> {coins}개")
print(f"100원짜리 ==> {coins1}개")
```
